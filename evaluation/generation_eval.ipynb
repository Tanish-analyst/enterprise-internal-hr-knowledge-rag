```python
import json
import time
import re
import cohere
from langchain_groq import ChatGroq

EVAL_FILE = "/content/generational_eval.jsonl"
MAX_QUERIES = 60

PINECONE_TOP_K = 10
PINECONE_SCORE_THRESHOLD = 0.5

groq_llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    api_key="gsk_........"
)

COHERE_TOP_K = 3
COHERE_SCORE_THRESHOLD = 0.5

COHERE_CALL_LIMIT = 10
COHERE_SLEEP_SECONDS = 60

co = cohere.ClientV2(api_key="Cy..........")
cohere_call_count = 0


def split_sentences(text: str):
    return [
        s.strip()
        for s in re.split(r'(?<=[.!?])\s+', text)
        if len(s.strip()) > 5
    ]


def judge_sentence_faithfulness(context: str, sentence: str, llm):
    prompt = f"""
You are evaluating faithfulness in a RAG system.

Context:
<<<
{context}
>>>

Answer sentence:
<<<
{sentence}
>>>

Is this sentence fully supported by the context?
Respond with only YES or NO.
"""
    resp = llm.invoke(prompt).content.strip().upper()
    return resp.startswith("YES")


def judge_answer_relevance(question: str, answer: str, llm):
    prompt = f"""
You are evaluating answer relevance in a RAG system.

Score the answer on a scale of 1 to 5:
5 = Fully answers the question precisely
4 = Answers the question but misses minor detail
3 = Partially answers the question
2 = Weakly related or generic
1 = Does not answer the question

Question:
<<<
{question}
>>>

Answer:
<<<
{answer}
>>>

Respond strictly in this format:
Score: <1-5>
Reason: <one short sentence>
"""
    resp = llm.invoke(prompt).content.strip()

    score = 1
    reason = "No reason returned"

    for line in resp.splitlines():
        if line.lower().startswith("score"):
            score = int(re.findall(r"\d+", line)[0])
        elif line.lower().startswith("reason"):
            reason = line.split(":", 1)[1].strip()

    return score, reason



eval_rows = []
with open(EVAL_FILE, "r") as f:
    for line in f:
        eval_rows.append(json.loads(line))

eval_rows = eval_rows[:MAX_QUERIES]


faithfulness_scores = []
relevance_scores = []


for idx, row in enumerate(eval_rows, 1):
    role = row["role"]
    question = row["question"]

    print(f"\n{'='*80}")
    print(f"[{idx}/{MAX_QUERIES}] ROLE={role} | Q={question}")
    print(f"{'='*80}")

    query_dense = emb.embed_query(question)
    query_sparse = bm25.encode_queries([question])[0]

    pinecone_res = index.query(
        vector=query_dense,
        sparse_vector=query_sparse,
        top_k=PINECONE_TOP_K,
        include_metadata=True,
        filter={role: {"$eq": True}}
    )

    pinecone_filtered = [
        m for m in pinecone_res.matches
        if (m.score or 0) >= PINECONE_SCORE_THRESHOLD
    ]

    final_docs = []

    if pinecone_filtered:
        doc_texts = [m.metadata.get("text", "") for m in pinecone_filtered]

        cohere_call_count += 1
        if cohere_call_count % COHERE_CALL_LIMIT == 0:
            print("â± Sleeping for Cohere free-tier rate limit...")
            time.sleep(COHERE_SLEEP_SECONDS)

        rer = co.rerank(
            model="rerank-v3.5",
            query=question,
            documents=doc_texts,
            top_n=len(doc_texts)
        )

        reranked = []
        for r in rer.results:
            match = pinecone_filtered[r.index]
            reranked.append({
                "text": match.metadata.get("text", ""),
                "pinecone_score": match.score,
                "cohere_score": r.relevance_score
            })

        reranked.sort(key=lambda x: x["cohere_score"], reverse=True)

        final_docs = [
            d for d in reranked[:COHERE_TOP_K]
            if d["cohere_score"] >= COHERE_SCORE_THRESHOLD
        ]

    if not final_docs:
        answer = (
            "The provided documents do not contain enough information "
            "to answer this question."
        )
        context = ""
        print("âŒ No documents passed filtering.")
    else:
        context = "\n\n---\n\n".join(d["text"] for d in final_docs)

        answer_prompt = f"""
You are an enterprise HR assistant.
Answer strictly using the provided context.
If the answer is not present, say you do not have enough information.

Context:
<<<
{context}
>>>

Question:
{question}
"""
        answer = groq_llm.invoke(answer_prompt).content.strip()

    print(f"\nğŸ§  Answer:\n{answer}")


    sentences = split_sentences(answer)

    supported = 0
    for s in sentences:
        if context and judge_sentence_faithfulness(context, s, groq_llm):
            supported += 1

    faithfulness = supported / max(len(sentences), 1)
    faithfulness_scores.append(faithfulness)

    print(f"\nğŸ“Š Faithfulness score: {faithfulness:.2f}")


    relevance_score, relevance_reason = judge_answer_relevance(
        question, answer, groq_llm
    )
    relevance_scores.append(relevance_score)

    print(f"ğŸ¯ Answer Relevance: {relevance_score} / 5")
    print(f"ğŸ“ Reason: {relevance_reason}")


print("\n" + "="*80)
print("ğŸ“Š FINAL EVALUATION RESULTS")
print("="*80)

avg_faithfulness = sum(faithfulness_scores) / len(faithfulness_scores)
avg_relevance = sum(relevance_scores) / len(relevance_scores)

print(f"Avg Faithfulness      : {avg_faithfulness:.3f}")
print(f"Avg Answer Relevance  : {avg_relevance:.2f} / 5")
print("="*80)
```
