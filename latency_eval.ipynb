```python
from pinecone import Pinecone, ServerlessSpec
import os
os.environ["PINECONE_API_KEY"] = "pcsk_........."

pc = Pinecone()

index_name = "final-rag-trailer"

index = pc.Index(index_name)
```

```python
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
from pinecone_text.sparse import BM25Encoder
bm25 = BM25Encoder.default()
```

```python
emb = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
import json

FILE_PATH = "/content/latency_eval.jsonl"

records = []

with open(FILE_PATH, "r") as f:
    for line in f:
        records.append(json.loads(line))

print(f"Total records: {len(records)}")

print("\nFirst 5 records:\n")
for i, record in enumerate(records[:5], start=1):
    print(f"Record {i}:")
    print(json.dumps(record, indent=2))
    print("-" * 50)
```

```python
import json
import time
import numpy as np
import cohere
from langchain_groq import ChatGroq

EVAL_FILE = "/content/latency_eval.jsonl"
MAX_QUERIES = 60

PINECONE_TOP_K = 10
PINECONE_SCORE_THRESHOLD = 0.5

COHERE_TOP_K = 3
COHERE_CALL_LIMIT = 10
COHERE_SLEEP_SECONDS = 60

groq_llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    api_key="gsk_......"
)

co = cohere.ClientV2(api_key="Cy......")

eval_rows = []
with open(EVAL_FILE, "r") as f:
    for line in f:
        eval_rows.append(json.loads(line))

eval_rows = eval_rows[:MAX_QUERIES]

print("🔥 Warming up LLM...")
for _ in range(3):
    groq_llm.invoke("Hello")

latencies = []
cohere_call_count = 0


for idx, row in enumerate(eval_rows, 1):

    role = row["role"]
    question = row["question"]

    print(f"\n[{idx}/{MAX_QUERIES}] ROLE={role} | Q={question}")


    if cohere_call_count > 0 and cohere_call_count % COHERE_CALL_LIMIT == 0:
        print("⏳ Sleeping due to Cohere free-tier limit...")
        time.sleep(COHERE_SLEEP_SECONDS)

    start = time.perf_counter()

    query_dense = emb.embed_query(question)
    query_sparse = bm25.encode_queries([question])[0]

    pinecone_res = index.query(
        vector=query_dense,
        sparse_vector=query_sparse,
        top_k=PINECONE_TOP_K,
        include_metadata=True,
        filter={role: {"$eq": True}}
    )

    pinecone_filtered = [
        m for m in pinecone_res.matches
        if (m.score or 0) >= PINECONE_SCORE_THRESHOLD
    ]

    final_docs = []

    if pinecone_filtered:
        doc_texts = [m.metadata.get("text", "") for m in pinecone_filtered]

        rer = co.rerank(
            model="rerank-v3.5",
            query=question,
            documents=doc_texts,
            top_n=len(doc_texts)
        )

        cohere_call_count += 1

        reranked = []
        for r in rer.results:
            match = pinecone_filtered[r.index]
            reranked.append({
                "text": match.metadata.get("text", ""),
                "cohere_score": r.relevance_score
            })

        reranked.sort(
            key=lambda x: x["cohere_score"],
            reverse=True
        )

        final_docs = reranked[:COHERE_TOP_K]

    if not final_docs:
        answer = "Not enough information."
    else:
        context = "\n\n---\n\n".join(d["text"] for d in final_docs)

        answer_prompt = f"""
You are an enterprise HR assistant.

STRICT RULES:
1. Use ONLY the information provided.
2. If missing, say you don't have enough information.
3. No assumptions.

Context:
<<<
{context}
>>>

Question:
{question}

Final Answer:
"""
        answer = groq_llm.invoke(answer_prompt).content.strip()

    end = time.perf_counter()

    latency = end - start
    latencies.append(latency)
    print(f"Latency: {latency:.2f}s")


lat = np.array(latencies)

print("\n" + "="*70)
print("🚀 END-TO-END LATENCY METRICS")
print("="*70)
print(f"P50 : {np.percentile(lat,50):.2f}s")
print(f"P95 : {np.percentile(lat,95):.2f}s")
print(f"P99 : {np.percentile(lat,99):.2f}s")
print("="*70)
```
